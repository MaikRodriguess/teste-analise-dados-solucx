### Importação de dados

```python
import pandas as pd
carros_df = pd.read_csv(r"auto.csv", sep = ";", encoding="latin1")
clientes_df = pd.read_csv(r"clientes1.csv", sep = ";", encoding="latin1")
carros_df  = carros_df.rename(columns={"ï»¿auto_id":"auto_id","id_cliente":"id_client"})
display(clientes_df)
display(carros_df)
```

### 2) Qual a média salarial dos clientes para cada estado do sudeste?							


```python
clientes_estados_sudoeste= clientes_df.loc[clientes_df["state"].isin(["SP", "RJ", "MG", "ES"])]
media_estado = clientes_estados_sudoeste.groupby("state").mean()
media_estado = media_estado[["monthly_income"]]
def formatar(valor):
    return f"${valor:,.2f}"
media_estado["monthly_income"] = media_estado["monthly_income"].map(formatar)
display(media_estado)
```

### 3) Qual o maior salário? E o menor?

```python
estado_salario = clientes_estados_sudoeste[["state", "monthly_income"]]
estado_salario = estado_salario.groupby("state").agg(['max', 'min'])
# estado_salario = clientes_df[["state", "monthly_income"]]
# estado_salario = estado_salario.groupby("state").agg(['min', 'max'])
display(estado_salario)
```

### 6) Quantas cidades distintas existem dentre os clientes do RS e quais são elas?
